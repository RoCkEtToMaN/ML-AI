##Liste

###Indexing

Ci permette di accedere ad un elemento specifico della lista.
Vediamo come si può fare:


```
print("Primo elemento della lista: ",l[0])     
print("Secondo elemento della lista: ",l[1])   
```

Possiamo anche puntare direttamente all'ultimo elemento della lista, penultimo...insomma andare a ritroso:

```
print("Ultimo elemnto della lista: ", l[-1])
print("Penultimo elemnto della lista: ", l[-2])
```

###Slicing

Un modo per listare una sottolista:

```
print(l[2:5])     #Listo dalla 3° incluso al 6° escluso
print(l[:3])      #Fino al 3°
print(l[-2:])     #Ultimi due elementi
print(l[:-2])     #FIno agli ultimi due elemnti, esclusi
```

```
print(l[::])   #Altro modo di stampare tutta la lista
print(l[::-1]) #Interessante modo di stampare la lista al contrario
```

Vediamo la modifica della lista:

```
l[4]=10
l
```

##Sets

Sono insiemi di elementi unici non ordinati, cioè non può contenere due volte uno stesso elemento e non tiene conto della disposizione degli elementi al suo interno.




```
sets={1,2,3,4,5,5}
print(sets,type(sets))
```

Vediamo alcune funzioni utili che ci permettono di lavorare con il set.



```
subjects_sets={"Italiano","Latino","Storia","Scienze","Filosofia","Italiano","Italiano"}
subjects_sets
```

Vediamo l' *Add*:

```
subjects_sets.add("Chimica")
subjects_sets
```

Vediamo come rispetta l'ordine alfabetico.


Vediamo il *Remove*:

```
subjects_sets.remove("Filosofia")
subjects_sets
```

Ovviamente, se proviamo a rimuovere un qualcosa che non c'è, avremo un eccezione.

```
subjects_sets.remove("Arte")
```

Se non vogliamo avere errore *Discard*:

```
subjects_sets.discard("Arte")
```

Se vogliamo estarre un elemente in modo random della mia lista, possiamo usare il metodo *Pop*:

```
subjetcs_pop=subjects_sets.pop()
print(subjetcs_pop,subjects_sets)
```

Se volessimo svuotare tutto il set dal suo contenuto, si usa *Clear*:



```
subjects_sets.clear()
print(subjects_sets)
```

Volendo possiamo creare un set "immutabile" ossia non lo si può più modificare,
si parlerà di un *Frozen Set*:

```
subjects_frozen= frozenset({"Italiano","Latino","Storia"})
print(subjects_frozen,type(subjects_frozen))
```

Proviamo a modifiare questo set:

```
subjects_frozen.add("CIAO")
```

Non si può modificare!


##I Dizionari

```
report={"Italiano":7,"Matematica":7,"Inglese":4,"Filosofia":6}
print(report,type(report))
```

```
report["Filosofia"]
```

```
report["Inglese"]
```

```
report["Filo"]
```

```
report["Elettronica"]=10
print(report)
```

Anche qui possiamo creare una collezione annidata:



```
report["Inglese"]={"grammatica":4,"Letteratura":7}
report
```

```
report["Inglese"]["Letteratura"]
```

Otteniamo solamente i valori del dizionario:

```
report.values()
```

Stessa cosa per le chiavi:

```
report.keys()
```

Con questi due oggetti non ce ne facciamo granchè, però si possono convertire in **Liste**:

```
keys=report.keys()
```

```
keys=list(keys)
print(keys,type(keys))
```

##Casting

Come abbiamo visto al rigo 39, si può convertire una collezione tra un tipo all'altro.
Vediamo come si può utilizzare il Casting.




```
names_list=["Giuseppe","Giacomo","Carlo","Antonio","Giacomo"]
print(names_list,type(names_list))
```

Convertiamo da Lista->Tupla

```
names_tuples=tuple(names_list)
print(names_tuples,type(names_tuples))
```

Proseguiamo    Tupla->Set:

```
names_set=set(names_tuples)
print(names_set,type(names_set))
names_set=set(names_list)
print(names_set,type(names_set))
```

Vediamo che la conversione in Set è valido sia da Tupla che da Lista e che gli elementi doppi sono stati tolti e l'ordine non è come quello di partenza, come i Set fanno.

Creiamo un dizionario partendo da una Tuple.

```
names_age=(("Giuseppe",10),("Federica",34),("Maria",33))
print(names_age,type(names_age))
people_dict=dict(names_age)
print(people_dict,type(people_dict))
```

Possiamo crare un dizionario partendo da due Liste:

```
names=["Giacomo","Andrea","Marco"]
ages=[24,56,11]

names_age=zip(names,ages)
people_dict=dict(names_age)
print(people_dict,type(people_dict))
```
