Queste operatori ci permettaranno di creare queste condizioni.



1.   "==": True quando le due espressioni sono identiche







```
1==1
```

```
"ciao"=="ciao"
```

```
"ciao"=="bao"
```

2. "!=": True quando le due espressioni sono diverse

```
"cane" != "gatto"
```

```
6.5>5.9
```

4. ">=": True se la prima espressione è maggiore e uguale alla seconda (solo per i numeri)

```
6>=6
```

5."<": True se la prima espressione è minore della seconda (solo per i numeri)

```
5<6
```

6."<=": True se la prima espressione è minore o uguale della seconda (solo per i numeri)

```
6<=6
```

```
"ad"<"ac"
```

##Statement IF-ELSE

Classico comando visto in diversi linguaggi di programmazione.


```
if(True):
    print("Questo codice viene eseguito")

if(False):
    print("Questo codice non viene eseguito")    
```

```
PRINT =True

if(PRINT == True):
    print("Questo testo viene stampato")
```

Esempio un po piu complesso:

```
EMAIL="giacomo.lo87@gmail.com"

email=input("inserisci l'email: ")

if(email==EMAIL):
    print("Hai 5 nuove email")
else:
    print("Accesso Negato")
```

Aggiungo un controllo: 

```
EMAIL="giacomo.lo87@gmail.com"
PASSWORD="ciao"
email=input("inserisci l'email: ")


if(email==EMAIL):
    password=input("inserisci la pass: ")
    if(password==PASSWORD):
        print("Hai 5 nuove email")
    else:
        print("La password è sbagliata. Accesso negato")
else:
    print("Indirizzo Inesistente")
```

##Operatori Logici



*   AND: ritorna TRUE se entrambe le condizioni sono True, altrimenti False






```
1==1 and "casa"!="albergo"
```

```
1==1 and "casa"!="casa"
```



*   OR: ritorna TRUE se le due condizioni sono diverse




```
1==2 or "casa"!="albergo"
```

```
1==2 or "casa"=="albergo"
```



*   NOT: esegue una negazione, se la condizione è True ritornerà False e viceversa




```
not(1==2 or "casa"!="albergo")
```

Vediamo il codice di prima modicato con gli *Operatori Logici*:

```
EMAIL="giacomo.lo87@gmail.com"
PASSWORD="ciao"
email=input("inserisci l'email: ")
password=input("inserisci la pass: ")

if(email==EMAIL and password == PASSWORD):    
    print("Hai 5 nuove email")
else:
    print("Accesso negato")
```

##Statement ELIF

Per condizioni Multiple.

Facciamo subito un esercizio dove inseriamo un numero e mettiamo dei controlli:


*   positivo, pari
*   positivo, dispari
*   negativo, pari
*   negativo, dispari






```
n = int(input("Inserire un numero :"))

if(n%2 == 0 and n>=0):
    print("%d è numero Positivo e Pari " %n )
elif(n%2 != 0 and n>=0):
    print("%d è numero Positivo e Dispari "%n)
elif(n%2 == 0 and n<0):
    print("%d è numero Negativo e Pari "%n)
elif(n%2 != 0 and n<0):
    print("%d è numero Negativo e Dispari "%n)
else:
    print("Abbiamo un problema!")
```

##Espressione IF

Questa espressione è diversa dallo statement IF, questa ci permette di impostare una valore di una variabile in funzione di una condizione.

Vediamo un esempio:
questo è con l'utilizzo del classico IF

```
n=int(input("Inserire un numero: "))

if(n%2==0):
    is_even=True
else:
    is_even=False
    
print("%d è pari? %s" %(n, is_even))    
```

Con l'Espressione IF:

```
n=int(input("Inserire un numero: "))

is_even= True if n%2==0 else False

print("%d è pari? %s" %(n, is_even))    
```

è molto utile nella funzione *Print* per mostrare un Output in maniera condizionale!

```
n=int(input("Inserire un numero: "))

print("%d è " %n, "pari"  if n%2 == 0 else "dispari")
```

Quindi abbiamo fatto tutto all'interno di un'unica espressione!
