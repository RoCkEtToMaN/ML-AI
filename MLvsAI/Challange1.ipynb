##**Challenge 1**

Inseririre due numeri interi come input e stampa tre linee dove:
1. La prima è la somma
2. Differenza 
3. Prodotto


```
a=int(input("Inserire il primo numero: "))     #Inseriamo i due numeri
b=int(input("Inserire il secondo numero: "))
```

```
sum=a+b;
print(f"La somma è: {sum}")
diff=a-b;
print(f"La somma è: {diff}")
molt=a*b;
print(f"La somma è: {molt}")
```

Eseguire lo Swapping: invertire il valore di due variabili.

1. Prendi in input due valori e assegnali a due variabili A e B

3. Stampa le variabili su schermo

```
A=input("Inserire il valore di A: ")    #Inseriamo le due variabili generiche
B=input("Inserire il valore di B: ")
```

```
C=A
A=B
B=C

print(f"Stampo la nuova A: {A}, stampo il nuovo B: {B} ")
```

```
A,B=B,A    #Altro Modo                                               
print(f"Stampo la nuova A: {A}, stampo il nuovo B: {B} ") 
```

#Casting e Formattazione

Scriviamo un programma che richieda i tuoi dati e li stampi su schermo, i dati che il programma chiede sono:

1. Nome e Cognome
2. Anno di nascita
3. Luogo di Nascita
4. Altezza (in metri)
5. Peso (in chilogrammi)

```
print("Inserire nome e cognome: ")    #Inserisco i dati
nome=input()
cognome=input()
print("Anno di nascita: ")
anno=int(input())
print("Luogo di nascita: ")
luogo=input()
print("Altezza (m): ")
altezza=float(input())
print("Peso (Kg): ")
peso=float(input())
```

```
print(f"{nome} {cognome}, nato a {luogo} nel {anno}, altezza: {altezza} m, peso: {peso} Kg") #Stampo i Dati 1° versione
```

```
print("%s %s, nato a %s nel %d, altezza: %.2f m, peso: %2.f kg" %(nome,cognome,luogo,anno,altezza,peso))  #Stampo i Dati 2° versione
```

```
print("{my_nome} {my_cognome}, nato a {my_luogo} nel {my_anno}, altezza: {my_altezza:.{altezza_digits}f} m, peso: {my_peso:.{peso_digits}} Kg"\ #Stampo i Dati 3° versione
      .format(my_nome=nome,my_cognome=cognome,my_luogo=luogo, my_anno=anno, my_altezza=altezza, altezza_digits=2, my_peso=peso, peso_digits=3))
```
