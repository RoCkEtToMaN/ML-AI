1.*Calcolo dell'area della circonferenza*:



```
def area_circo(r):
    return 3.14*r*r;    
```

```
r=float(input("Inserire il raggio della circonferenza: "))
area=area_circo(r)

print("L'area della circo di raggio %.2f è di %.2f " % (r,area))
```




2.   *Calcolo Anni Bisestili*:
     ogni 4 anni si aggiunge un giorno in più al mese più breve (febbraio).

     Se:


*   è divisibile per 4, allora è bisestile
*   * se è divisibile per 100, allora NO a meno che non sia 
* * * divisibile per 400





```
def is_bis(year):
    bis= False

    if(year%4==0):
        if(year%100==0):
            if(year%400==0):
              bis= True
        else:
            bis= True  


    return bis    

```

```
year=int(input("Inserire l'anno: "))
print("L'anno %d è bisestile = %s " % (year, is_bis(year)))
```
