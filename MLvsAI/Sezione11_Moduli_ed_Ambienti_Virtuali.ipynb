#***Sezione11: Moduli ed Ambienti Virtuali***

##***Intro***

Oltre ai moduli delle Standard Library, vi sono a disposizione moltri altri sviluppati da altri.

Andando sul sito  [https://pypi.org/](https://), possiamo trovare tutti i moduli di Python dati da Python Package Index!
All'interno al sito, i moduli sono raggruppati in diversi gruppi e sottogruppi.

Si trova il modulo interessato e si sceglie: vedremo al suo interno tutta la descrizione di quello che fanno e su come si installano.

##***Installare pip***

Per installare un modulo possiamo usare un software apposito chiamato *Pip*.

Per utilizzare basta lanciare dal cmd:

````
pip install modulo
````

Ma se è installato Python, Pyp è già presente ma in caso contrario:

````
pip3
````

Se non è presente cmq, scarichiamo il pacchetto e lo installiamo manualmente:

[https://bootstrap.pypa.io/get-pip.py](https://)



Si salva la finestra in formato py e nel caso fossimo in ambiente Linux, da terminale:



```
wget indirizzo
```

Successivamente, per entrambe le procedure :



```
python get-pip.py
```

Ma solitamente, pip viene installato in automatico con Python!




##***Installare pip il modulo***

Adesso procediamo con l'installazione del modulo:



```
pip install nomemodulo
```

Se lo volessi rimuovere:



```
pip uninstall nomemodulo
```

Se la reinstallazione non va a buon fine, questo potrebbe essere portato alla presenza di Anaconda ed utilizzando il prompt di Anaconda questo dovrebbe andare bene.





##***Utilizzare il modulo***

Una volta installato, per utilizzarlo:



```
import modulo
```



##***Creare Ambienti Virtuali***


Installando moduli esterni possiamo incontrare diverse problematiche:


*   Incompatibilità tra loro
*   occupazione spazio

Quindi la cosa piu intelligente è quello di installarli in appositi ambienti virtuali creati ad hoc per il progetto che stiamo maneggiando.

Cosa si utilizza?

##***Virtual Env***

QUesto ci permette di creare ambienti virtuali che si utilizzano per i vari progetti singolarmente.

Dal terminale di Anaconda, sfruttiamo *pip*:

```
pip install virtualenv
```

Poi per creare la cartella dell'apposito progetto:



```
mkdir my_project
```
Successivamente si creerà la cartella e ci sposteremo all'interno.

Per creare l'ambiente in quel path:



```
virtualenv nomeprogetto
```

Successivamente, vedremo che nel path si creerà una sottocartella con il nome del progetto e successivamente per questo progetto verranno installati i moduli che vorremo installare.

Per attivare l'ambiente virtuale da cmd:



```
nomeprogetto\Scripts\activate
```

Alla sinistra del terminale verrà rappresentata sulla dx:

(nomeprogetto)  path


Da Linux\Mac:



```
source nomeprogetto\bin\activate
```

Una volta che l'ambiente è avviato, installiamoci i moduli:



```
pip install numpy
```

Questo modulo lo utilizzeremo nei prossimi video!
Nella cartella Lib troveremo questo modulo installato!


*Provare con Jupyter!*


Creiamo il "main.py" dentro alla cartella del progetto e lo eseguiamo!

Per vedere se funziona, impostiamo questo:



```
import numpy as np

print(np.__version__)
```

Se lanciandolo da teminale: 


```
python main.py
```

Riceviamo la versione di numpy, il tutto funziona!

Per uscire dall'ambiente virtuale:



```
deactivate
```


Ad esempio:



*   *Pandas* : utile per l'analisi dati
*   *Scikit-Learn*: utile per il ML

Poi per vedere quali moduli abbiamo installato nel nostro ambiente virtuale:


```
pip freeze
```
Ovviamente con i moduli installati, abbiamo anche tutte le dipendenze.


Possiamo esportare i moduli che usiamo in questo progetto in un file:

```
pip freeze > requirements.txt

```

Poi chiudiamo il progetto (deactivate....) 

Creiamo un altro ambiente virtuale ed andando ad installare il file txt:

```
pip install -r requirement 
```

In questo modo, pip installa tutti i moduli contenuti nel file.


Sfruttando quindi la versatilità degli ambienti virtuali, è possibile lavorare su progetti di grandi dimensioni!



E alla fine possiamo disattivarlo!

##***Virtual Env con Anaconda***

Vediamo come gestire un ambiente virtuale con Anaconda stesso!



1.   *Prompt di Anaconda*:



```
conda create --name nomeambiente
```
  E successivamente, l'ambiente verrà creato.

2.   *Visualizziamo gli ambienti già presenti*:



```
conda env list
```



3.   *Attivare ambiente*:


```
conda activate env
```


4.   *Installare Moduli*: anche qui che verrano utilizzate con questo ambiente virtuale



```
conda install nomemodulo
```





1.   *Lista di moduli installati*:


```
conda list --name base
```
dove "base" è l'ambiente standard ma possiamo mettere i nomi di tutti gli ambienti che andremo a creare

2.   *Associare Jupyter Notebook*: possiamo dedicarlo per il nostro ambiente virtuale



```
conda install -n env nb_conda_kernels
```
Installa i moduli per Jupyter

Per installarlo veramente:



```
python -m ipykernel install --user --name nomekernel
```

Adesso possiamo avviare il Notebook:



```
jupyter notebook
```

Poi andremo a selezionare manualmente il kernel creato!



1.   *disattivare*:



```
conda deactivate
```


2.   *rimuovere l'ambiente*:



```
conda remove --name all nomeambiente -all
```


Cosi l'ambiente verrà disinstallato!
