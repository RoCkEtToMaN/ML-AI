##Variabili: i suoi tipi

```
num=5;
print(num)
```

```
word="ciao"
print(word)
type(word)
```

Vediamo che in Python le variabili non sono Tipizzate, ossia possiamo associare diversi tipi di valori alla stessa variabile

```
var=5  #Intero
var=4.15 #Float
var="ciao ciao"  #stringa
var=True  #Booleano


print(var,type(var))
```

Vedremo...anche le Classi quando inizieremo la prog. ad Oggetti

Se volessimo convertire un tipo di dato in un altro tipo di dato, lo si può fare utilizzando Casting: ossia convertire direttamente un tipo di variabile ad un altra.


- Da Intero a Float:

```
var=1
var=float(var)
print(var,type(var))
```

- Da Float ad Intero:

```
var=1.5
var=int(var)

print(var,type(var))
```

- Da Intero a Stringa

```
var=5
var=str(var)

print(var,type(var))
```

- Da Stringa  a Numero

```
var="5.2"
var=float(var)

print(var,type(var))
```

##Formattazione

```
age=25
print("Hai %d anni" %age)
```

- **%d**: numero intero
- **%f**: numero con virgola mobile
- **%s**: stringa

Comunque sul sito di Python ci sono le doc dove si possono trovare i parametri per la formattazioni

```
my_name=input("Come ti chiami?")
my_age=input("Quanti anni hai?")
my_age=int(my_age)
my_born= 2020-my_age

print("Il mio nome è %s " %my_name)
print("sono nato nel %d " %my_born)
print("Ti chiami %s ed hai %d anni e sei nato nel %d" %(my_name,my_age,my_born))
```

Altro metodo, utilizzando P3, è il metodo Format:

```
print("Il tuo nome è {name}, hai {age} anni, quindi sei nato nel {born}".format(name=my_name, age=my_age, born=my_born))
```

Se utilizziamo una versione >3.6, vi è una tecnica chiamata "fstream": metto la **f** subito dentro al 

```
print(f"il tuo nome e' {my_name}, hai {my_age} e sei nato nel {my_born}")
```
