#Sezione 5: ***La Programmazione Procedurale e le Funzioni***

è un paradigma di programmazione la cui la logica del programma viene valutata dal *basso verso l'alto*.

Il vantaggio di questo tipo sono le **Procedure/Funzioni/Routine**, questi sono blocchi di codice che svolgono un determinato task e sono riutilizzabili tramite degli Alias che assegnamo noi.

Ad esempio, consideriamo una funzione per poter calcolare la lunghezza di una lista (Python da già a disposizione una sua funzione ma proviamo ad implementarla "*len*").



```
l=["ciao","fame","corsa","sbatta"];
print(l);

count=0;

for i in l:
  count=count+1;


print("Nella lista ci sono %d elementi" %count);
```

Classico modo per ottenere il numero degli elementi contenuti.

Ma se volessimo utilizzare le funzioni? o meglio se volessi usare questo algoritmo in piu parti del codice?

Dovremmo andare a definire questo algoritmo diverse volte in funzione delle liste che si possono incontrare durante la scrittura di un file.

QUindi la soluzione è definire l'algoritmo all'interno della funzione.





```
def list_len(l):

    count = 0
    for i in l:
        count+=1;

    return count


l_1=["ciao","mamma"];
print("Nella lista 1 ci sono %d elementi" %list_len(l_1));

l_2=["mia","mavaaaa","richthhhha"];
print("Nella lista 2 ci sono %d elementi" %list_len(l_2));

l_3=["boxe","pugile","ballerino","loop","cambiare lavoro"];
print("Nella lista 3 ci sono %d elementi" %list_len(l_3));
```

##Esempio: Area Tringolo

Vediamo come creare una funzione per il calcolo di area di un triangolo:

```
def area_triangolo(b,h):
    return (b*h)/2

    
```

```
area_triangolo(3,4)
```

```
b=float(input("Inserisci la base: "))
h=float(input("Inserisci l'altezza: "))
area=area_triangolo(b,h)
print("Area del triangolo di base %.2f e altezza %.2f è pari a %.2f" % (b,h,area))
```

##Esempio: Lista della Spesa

Vediamo come creare una funzione per stampare una lista della spesa passata in input:

```
def print_list_spesa(lista, owner="Giacomo"):

    print("La lista della spesa di "+owner)

    for i,entry in enumerate(lista):
        print("%d) %s " % (i+1,entry))


```

```
lista=["riso","birra","focaccia"]

print_list_spesa(lista)
```

Vediamo come possiamo giocare con l'aggiunta di diversi argomenti all'interno della funzione:

```
lista=["riso","birra","focaccia","Caramelle","Tonno","ricotta"]
print_list_spesa(lista,"Carlo")
```
